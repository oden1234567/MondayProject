```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

1.Matplot basic visualization

```python
#line plot
x = np.linspace(0,10,100)
y =np.tan(x)
```

```python
print(x)
print("__________")
print(y)
```

```python
plt.plot(x,y,label="sinusodial wave")
plt.title("Line plot example graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
#Bar chart
categories = ["A","B","C","D"]
values = [1,4,2,3]
plt.bar("A",3,color="violet")
plt.bar("B",4,color="red")
plt.bar("C",2,color="blue")
plt.bar("D",0.25,color="green")
plt.title("Bar chart example graph")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python

#Histogram and KDE plot
data = np.random.randn(1000)
sns.histplot(data,kde=True,color="red")
plt.title("Histogram and KDE plot example graph")
plt.xlabel("Data")
plt.ylabel("Frequency")
plt.show()
```

```python
#Heatmap
matrix = np.random.rand(4,4)
sns.heatmap(matrix,annot=True,cmap="viridis",fmt=".2f")
plt.title("Heatmap example graph")
plt.show()
```

```python
#pairplot
sns.set_style("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df,hue="species",palette="viridis")
plt.title("Pairplot example graph")
```
